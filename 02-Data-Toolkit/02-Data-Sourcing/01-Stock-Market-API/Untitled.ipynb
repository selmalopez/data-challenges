## Import

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
```

```python
import requests

url = "https://iex.lewagon.com/stable/stock/aapl/chart/3m"
api_data = requests.get(url).json()
apple_df = pd.DataFrame(api_data)
apple_df['date'] = pd.to_datetime(apple_df['date'], format="%Y-%m-%d")
apple_df = apple_df.set_index('date')
```

```python
apple_df['close'].plot()
```

```python
apple_df['open'].plot()
```

```python
apple_df['low'].plot()
```

```python
apple_df['high'].plot()
```

```python
apple_df[['open', 'close', 'high', 'low']].plot(figsize=(12,4))
```
